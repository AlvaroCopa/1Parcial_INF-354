>Metodos de la Media, Moda y DesviaciÃ³n Estandar

```python
from math import sqrt
def media(v):
    s = 0
    for elem in v:
        if elem != None:
            s+=elem
    if len(v) % 2==0:
        p=len(v)
    else:
        p=len(v)+1
    return(s/p)

def moda(v):
    m = []
    cont = 0
    for elem in v:
        if elem != None:
            if v.count(elem) > cont:
                m=[]
                m.append(elem)
                cont = v.count(elem)
            elif v.count(elem) == cont:
                if not elem in m:
                    m.append(elem)
                    cont = v.count(elem)
    return(sorted(m))

def desviacionEstandar(v):
    med = media(v)
    s = 0
    for elem in v:
        if elem != None:
            s+= (elem - med)**2
    res = sqrt(s/(len(v)-1))
    return(res)
```

>DATASET drinking water potability

```python
import csv
matriz = []
with open('wine.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    for fila in reader:
        matriz.append(fila)
i = 0
for el in matriz[0]:
    x = [m[i] for m in matriz[1:]]
    v=[]
    for a in x:
        if "" != a:
            v.append(float(a))
        else: 
            v.append(None)
    print("===============================", matriz[0][i] ,"===============================")
    print("La Media es:", media(v))
    aux=moda(v)
    if len(aux)>3:
        print("La Moda es: ",end="" )
        print(*aux[0:3],sep=",",end=",....," )
        print(*aux[-3:],sep=",")
        print("\tTamaÃ±o del vector moda:",len(aux) )
    else:
        print("La Moda es: ", moda(v))
    print("La DesviaciÃ³n Estandar es:", desviacionEstandar(v))
    print("=================================================================================")
    i+=1
```

```python
import pandas as pd
data = pd.read_csv("wine.csv",header=[0])
for i in data:
    print("===============================", i ,"===============================")
    print("La Media es:", data[i].mean())
    print("La Moda es:", data[i].mode())
    print("La DesviaciÃ³n Estandar es:", data[i].std())
    print("=================================================================================")
```
